### .loc in pandas

```python
# In Pandas, .loc is a label-based data selection method.
#It lets you access rows and columns using names (labels) instead of positions.
```

```python
# returns all columns for row 'A'
newdf.loc["A"]
```

```python
newdf.loc[["A", "C", "E"]]
# Returns only rows "A", "C", and "E".
```

```python
newdf.loc["B", "D"]
# Returns the value from row "B" and column "D" (0.789306 in your table).
```

```python
newdf.loc[["B", "E", "H"], ["B", "D"]]
# Returns only the "B" and "D" columns for rows "B", "E", "H".
```

```python
newdf.loc["B":"E"]
# Returns all columns for rows "B" to "E".
```

```python
newdf.loc[newdf["C"] > 0.5]
# Returns all columns for rows where the value in column "C" is greater than 0.5.
```

```python
newdf.loc["A", "A"] = 1.0
# Sets the value at row "A" and column "A" to 1.0.
newdf
```

### .iloc in pandas

```python
# In Pandas, .iloc is the integer position–based selection method.
# It works like .loc, but instead of using labels, it uses row and column positions starting from 0.
```

```python
newdf.iloc[0]   # First row (label "A")
# returns zeroth row 
```

```python
# select multiple rows by position
newdf.iloc[[0, 2, 4]]   # Rows "A", "C", "E"
```

```python
# select a single value by positon 
newdf.iloc[1, 3]   # Row index 1 ("B"), column index 3 ("D") → 0.789306
```

```python
# select specific rows and columns by position
newdf.iloc[[1, 4, 7], [1, 3]]   # Rows B, E, H → Columns B, D
```

```python
# select a range of row  
newdf.iloc[1:4, 0:3]   # Rows B → D, Columns A → C
```

```python
# update a value 
newdf.iloc[0, 0] = 1.0   # First row, first column
newdf
```

#### .loc → label-based indexing (uses names like "A").
#### .iloc → integer position–based indexing (uses numbers like 0, 1, 2...).

## Drop row & column

```python
newdf.drop(['A' , 'C'] , axis=1 , inplace=False)  # Drop both 'A' and 'C' columns
```
